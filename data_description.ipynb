```python
import pandas as pd

# CSV-Datei laden
df = pd.read_csv(r"C:\Users\leona\Downloads\used_car_sales.csv")

# Erste Zeilen anzeigen
df.head()
```

```python
# Statistische Übersicht
print("Statistik für 'Sold Price-$':")
print(df["Sold Price-$"].describe())

# Fehlende Werte prüfen
print("\nFehlende Werte:", df["Sold Price-$"].isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df["Sold Price-$"], bins=20, edgecolor='black')
plt.title("Verteilung des Verkaufspreises")
plt.xlabel("Sold Price-$")
plt.ylabel("Anzahl")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["Mileage-KM"], df["Sold Price-$"], alpha=0.5)
plt.title("Verkaufspreis vs. Kilometerstand")
plt.xlabel("Mileage-KM")
plt.ylabel("Sold Price-$")
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df["Sold Price-$ (scaled)"] = scaler.fit_transform(df[["Sold Price-$"]])

# Überprüfen
df[["Sold Price-$", "Sold Price-$ (scaled)"]].head()
```

Ich habe das Feld **Sold Price-$** gewählt, da es ein zentraler Kennwert im Autoverkauf ist und sich gut für Vorhersagemodelle eignet. Die Analyse zeigt eine starke Streuung der Werte, was verschiedene Fahrzeugarten und Zustände widerspiegelt. Einige Einträge mit einem Verkaufspreis von 0 könnten auf unverkaufte oder falsch erfasste Fahrzeuge hinweisen. Insgesamt ist das Feld aber geeignet für weitere statistische Auswertungen und maschinelles Lernen.
